### Filtering based on a criteria

Untuk memfilter dataframe berdasarkan kondisi, kita cukup menggunakan kondisi boolean.

Contoh, filter dataframe untuk data platform pada pc(Platform = PC).


```python
# 1. create the condition boolean series
cond = df['Platform'] == 'PC'
print(cond)
```

```python
# 2. pass this condition into the dataFrame
df[cond]
```

```python
# we can pass multiple criteria in the same line.
# let's filter for Platform = PC and Year of Release = 2013

df[(df['Platform']=='PC') & (df['Year of Release']== 2013)]

# note: When using 'and' and 'or' operators, pandas requires we use '&' and '|' instead of 'and' and 'or'
# don't forget to enclose the two conditions in parentheses
```

# Line Pots (Series/Dataframe) <a id="12"></a>


Kita akan mendapatkan perbedaan sales di semua region untuk game GTA V pada platform PC

```python
sales = ['NA Sales', 'EU Sales', 'JP Sales', 'Other Sales', 'Global Sales']

selectedGame = df.loc['Grand Theft Auto V']
forPloting = selectedGame[selectedGame['Platform'] == 'PC']

result = forPloting.loc['Grand Theft Auto V', sales]
result.head()
```

Selanjutnya, membuat plot line dengan appending `.plot()` pada `Grand Theft Auto V`.

```python
result.plot()
```

*pandas* automatically populated the x-axis with the index values (years), and the y-axis with the column values (population). However, notice how the years were not displayed because they are of type *string*. Therefore, let's change the type of the index values to *integer* for plotting.

Also, let's label the x and y axis using `plt.title()`, `plt.ylabel()`, and `plt.xlabel()` as follows:


```python
result.plot(kind = 'line')

plt.title('Penjualan game GTA V di berbagai region')
plt.ylabel('Indeks')
plt.xlabel('Region')

plt.show()
```
